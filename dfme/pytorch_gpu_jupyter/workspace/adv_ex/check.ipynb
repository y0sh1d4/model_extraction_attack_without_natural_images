```python
ms = np.load(os.path.join('.', 'checkpoint', 'mnist', 'student_small', 'result.npz'))
mm = np.load(os.path.join('.', 'checkpoint', 'mnist', 'student_medium', 'result.npz'))
ml = np.load(os.path.join('.', 'checkpoint', 'mnist', 'student_large', 'result.npz'))

fs = np.load(os.path.join('.', 'checkpoint', 'fashion_mnist', 'student_small', 'result.npz'))
fm = np.load(os.path.join('.', 'checkpoint', 'fashion_mnist', 'student_medium', 'result.npz'))
fl = np.load(os.path.join('.', 'checkpoint', 'fashion_mnist', 'student_large', 'result.npz'))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(ms['acc'], label='small')
ax.plot(mm['acc'], label='medium')
ax.plot(ml['acc'], label='large')
ax.legend()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(fs['acc'], label='small')
ax.plot(fm['acc'], label='medium')
ax.plot(fl['acc'], label='large')
ax.legend()
plt.show()
```

```python
print('mnist/small', np.max(ms['acc']))
print('mnist/medium', np.max(mm['acc']))
print('mnist/large', np.max(ml['acc']))

print('fashion_mnist/small', np.max(fs['acc']))
print('fashion_mnist/medium', np.max(fm['acc']))
print('fashion_mnist/large', np.max(fl['acc']))
```
