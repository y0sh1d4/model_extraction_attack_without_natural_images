```python
import os
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 12
```

```python
mlflow_folder = os.path.join('/', 'data', 'mlruns', '3')
```

```python
# params.victim.model=medium, params.victim.task_name=mnist
mnist = {
    'small': 'd913e2bccde54ddc89ebcd5ff7848387',
    'medium': '5d05665b5a9643b58597e65ace1b4633',
    'large': 'daec10e142464fc28bcfe6d8927de21c',
}

fashion_mnist = {
    'small': 'c6153eb7a39447cbaae09c168a6bfbf0',
    'medium': '4d1a127ebf4044e08345b4d5a3071b90',
    'large': '701fee67b09747e5a3850c17c2d6b393',
}

fractal_1k = {
    'small': '7b67fb7ecc194e4fb36c3c7aa1e6018d',
    'medium': '43f3677d77e74a2b88d6cd558d0366e3',
    'large': 'a529b4d66c80426494a911d8324126eb',
}

fractal_60 = {
    'small': '1ea1f92c35444456b326d57788b185e1',
    'medium': '87cc72ffe8204bb3a56641da8d3c6dfe',
    'large': 'e26f84ed79334e35a773c7ccc121fa6f',
}
```

```python
attack_model = 'medium'
task_name = 'mnist'
colors = ['tab:blue', 'tab:orange', 'tab:green']
eps = '16/255.0'
fig = plt.figure()
ax = fig.add_subplot(111)

p = os.path.join(mlflow_folder, mnist[attack_model], 'artifacts', 'success_rate.npz')
sr = np.load(p)[eps]
ax.plot(list(range(1, len(sr)+1)), sr, label='$D^{EX}$=mnist')

p = os.path.join(mlflow_folder, fashion_mnist[attack_model], 'artifacts', 'success_rate.npz')
sr = np.load(p)[eps]
ax.plot(list(range(1, len(sr)+1)), sr, label='$D^{EX}$=fashion_mnist')

p = os.path.join(mlflow_folder, fractal_1k[attack_model], 'artifacts', 'success_rate.npz')
sr = np.load(p)[eps]
ax.plot(list(range(1, len(sr)+1)), sr, label='$D^{EX}$=fractalDB_1k')

p = os.path.join(mlflow_folder, fractal_60[attack_model], 'artifacts', 'success_rate.npz')
sr = np.load(p)[eps]
ax.plot(list(range(1, len(sr)+1)), sr, label='$D^{EX}$=fractalDB_60')

ax.legend()
plt.show()

fig.savefig(f'./figures/sr_{task_name}_{attack_model}.png', dpi=300, bbox_inches='tight', pad_inches=0)
```
