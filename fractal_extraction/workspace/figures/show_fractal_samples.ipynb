Show FractalDB samples

```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
ds_ext = tf.keras.utils.image_dataset_from_directory(
    directory='/dataset/fractal_pub/fractaldb_cat60_ins1000',
    labels='inferred',
    label_mode='int',
    color_mode='grayscale',
    shuffle=False,
    batch_size=1000,
)
```

```python
samples = []
for batch in ds_ext:
    samples.append(batch[0].numpy())
```

```python
w = 10
h = 3
n = 0
fig = plt.figure(figsize=(w, h))
for i in range(h):
    for j in range(w):
        ax = fig.add_subplot(h,w,(j+1)+w*i)
        ax.imshow(samples[j+w*i][n, :, :, 0], 'gray')
        ax.axis('off')
fig.tight_layout()
plt.show()
fig.savefig('./figures/fractal_samples_raw.png', dpi=300, bbox_inches='tight', pad_inches=0)
```

```python
w = 10
h = 3
n = 0
fig = plt.figure(figsize=(w, h))
for i in range(h):
    for j in range(w):
        ax = fig.add_subplot(h,w,(j+1)+w*i)
        ax.imshow(tf.image.resize(samples[j+w*i], (28,28))[n, :, :, 0], 'gray')
        ax.axis('off')
fig.tight_layout()
plt.show()
fig.savefig('./figures/fractal_samples_scaled.png', dpi=300, bbox_inches='tight', pad_inches=0)
```
